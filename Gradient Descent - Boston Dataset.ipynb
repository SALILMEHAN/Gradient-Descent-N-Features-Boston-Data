```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("Gradient Descent - Boston Dataset_train.csv")
data
```

```python
df=data.copy()
df.columns=['x1','x2','x3','x4','x5','x6','x7','x8','x9','x10','x11','x12','x13','y']
df
```

```python
def gd_step(df,alpha,m,c):
    m_new=[]
    c_new=0
    w=np.array([0 for i in range (379)])
    for i in range(df.shape[1]-2):
        w=w+(m[i]*df[df.columns[i]])
    dcost_by_dmi=df.y-w-c
    for i in range(12):
        t=m[i]-alpha*(-2/df.shape[0])*((dcost_by_dmi)*df[df.columns[i]]).sum()
        m_new.append(t)
    c_new=c-alpha*(-2/df.shape[0])*(dcost_by_dmi).sum()
    return m_new,c_new
```

```python
def cost(df,alpha,m,c):
    w=np.array([0 for i in range (379)])
    for i in range(df.shape[1]-2):
        w=w+(m[i]*df[df.columns[i]])
    t=((df.y-w-c)**2).sum()/df.shape[0]
    print(t)
```

```python
def gd(df,iterations,alpha):
    m=np.array([0 for i in range(12)])
    c=0
    for i in range(iterations):
        m,c=gd_step(df,alpha,m,c)
#         cost(df,alpha,m,c)
    return m,c
```

```python
def run():
    iterations=1000
    alpha=0.175
    m,c=gd(df,iterations,alpha)
    return m,c
```

```python
m,c=run()
```
